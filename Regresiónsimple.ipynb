```python
# prompt: realiza un analisis de regresion simple, teniendo en cuenta que la variable independiente Horas_Trabajadas y la variable dependiente Productos_Terminados y gbenerar un grafico con la linea de tendencia. Necesitan determinar el ceoficiente de determinacion. Necesito que generes una prediccion para un valor de X de 35 Horas_Trabajadas

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load the Excel file into a pandas DataFrame.
data = pd.read_excel("/content/drive/MyDrive/Estadística Industrial/DATOS/datos_regresion_simple.xlsx")

# Prepare the data
X = data['Horas_Trabajadas'].values.reshape(-1, 1)  # Reshape for sklearn
y = data['Productos_Terminados'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Calculate the coefficient of determination (R-squared)
r_squared = model.score(X, y)

# Predict for a new value of X (35 hours)
new_X = np.array([[35]])
predicted_y = model.predict(new_X)

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X, y_pred, color='red', label='Linea de tendencia')
plt.xlabel('Horas Trabajadas')
plt.ylabel('Productos Terminados')
plt.title('Regresión Lineal Simple')
plt.text(10, 100, f'$R^2$ = {r_squared:.2f}', fontsize=12)  # Display R-squared on the plot
plt.legend()
plt.grid(True)
plt.show()

print(f"Predicción para 35 horas trabajadas: {predicted_y[0]:.2f} productos terminados")
print(f"Coeficiente de determinación (R^2): {r_squared:.2f}")
```
